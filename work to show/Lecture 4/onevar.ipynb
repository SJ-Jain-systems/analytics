```python
import pandas as pd
import numpy as np
import seaborn as sns
df = pd.read_csv("iowa.csv",  encoding = 'latin1')
df.head()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

df["Sale_ln"] = np.nan
df.loc[df["Sale (Dollars)"] > 0, "Sale_ln"] = np.log(df.loc[df["Sale (Dollars)"] > 0, "Sale (Dollars)"])

df["Sale_ihs"] = np.arcsinh(df["Sale (Dollars)"])
df_kde = df[df["Sale (Dollars)"] > 0]

plt.figure()
sns.kdeplot(
    data=df_kde,
    x="Sale (Dollars)"
)
plt.title("KDE of Sale (Dollars)")
plt.xlabel("Sale (Dollars)")
plt.show()

plt.figure()
sns.kdeplot(data=df_kde, x="Sale_ln")
plt.title("KDE of log(Sale Dollars)")
plt.xlabel("log(Sale Dollars)")
plt.show()

plt.figure()
sns.kdeplot(data=df,x="Sale_ihs")
plt.title("KDE")
plt.xlabel("IHS(Sale Dollars)")
plt.show()
```
